### Source NASA dataset but retaining the datapoints by day and creates a csv for multiple visualizations. 
* This notebook takes the origianl source from NASA and creates a file that can be used for other visualizations. 

```python
%matplotlib inline 

import requests 
from io import StringIO
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import numpy.linalg as lin
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
sea_data = pd.read_csv('data/GMSL_merged_nasa_1993_2020.csv')
# sea_data 
sea_data.head(20)

# sea_level = sea_data['GMSL'].values
# sea_level
```

```python
from datetime import timedelta, datetime

```

```python
def convert_partial_year(number):
    year = int(number)
    d = timedelta(days=(number - year)*365.25)
    day_one = datetime(year,1,1)
    date = d + day_one
    return date
```

```python
sea_data['date'] = sea_data['year_fraction'].apply(convert_partial_year)
sea_data
```

```python
sea_data_1 = sea_data.drop(['year_fraction'], axis=1)
sea_data_1
```

```python
sea_data_1['year'] = sea_data_1['date'].apply(lambda x: x.year)
sea_data_1
```

```python
sea_data_2 = sea_data_1.drop(['date'], axis=1)
sea_data_2
```

```python
sea_data_2.columns = ['GMSL', 'Year']
sea_data_2
```

```python
#Saving the file that helps make visualizations to represent all of the data by day and all inclusive
sea_data_2.to_csv("data/sealevel_data__all_1993-2020.csv", index = False, header = True)
```
